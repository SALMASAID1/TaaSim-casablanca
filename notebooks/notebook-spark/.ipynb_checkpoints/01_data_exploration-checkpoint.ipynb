# Initialize the SparkSession 

```python
spark = SparkSession.builder \
    .appName("PortoTripsAnalysis") \
    .getOrCreate()
```

```python
from pyspark.sql import Row
from datetime import datetime

# 1. Create a tiny test dataset
test_data = [
    Row(service="Spark-to-MinIO", status="Connected", timestamp=datetime.now()),
    Row(service="S3A-Connector", status="Operational", timestamp=datetime.now())
]
test_df = spark.createDataFrame(test_data)

# 2. Define the path (using the bucket created in your minio-init service)
write_path = "s3a://taasim/raw/"

# 3. Write the data
print(f"Writing test data to {write_path}...")
test_df.write.mode("overwrite").parquet(write_path)
print("✅ Write Successful!")
```
